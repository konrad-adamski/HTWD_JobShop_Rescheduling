```python
def compare_machine_sequences_after_T1(prev_plan: pd.DataFrame,
                                       revised_plan: pd.DataFrame,
                                       T1: float) -> pd.DataFrame:
    """
    Vergleicht die Job-Reihenfolge je Maschine nach T1.

    Spalten:
    - original_sequence      : Reihenfolge der Jobs im Ursprungsplan (nach T1)
    - revised_sequence_raw   : Reihenfolge im Rescheduling-Plan (nach T1, ungefiltert)
    - revised_sequence       : revised_sequence_raw gefiltert auf original_sequence
    - changed                : True, wenn Reihenfolge geändert wurde
    """
    # Einheitliche Spalten
    df_prev = prev_plan.rename(columns={'Start': 'Start_prev', 'Job': 'Job'})
    df_rev  = revised_plan.rename(columns={'Start': 'Start_rev', 'ob': 'Job'})

    # Filter nach T1
    df_prev = df_prev[df_prev['Start_prev'] >= T1]
    df_rev  = df_rev[df_rev['Start_rev'] >= T1]

    # Sortieren
    df_prev_sel = df_prev[['Machine', 'Start_prev', 'Job']].sort_values(['Machine', 'Start_prev'])
    df_rev_sel  = df_rev[['Machine', 'Start_rev', 'Job']].sort_values(['Machine', 'Start_rev'])

    # Gruppieren
    original_sequence = df_prev_sel.groupby('Machine')['Job'].apply(list)
    revised_sequence_raw = df_rev_sel.groupby('Machine')['Job'].apply(list)

    # Vergleichstabelle
    comparison = pd.DataFrame({
        'original_sequence': original_sequence,
        'revised_sequence_raw': revised_sequence_raw
    }).dropna()

    # Gefilterte revised_sequence
    revised_filtered = {
        machine: [job for job in revised_sequence_raw.get(machine, []) if job in original_sequence[machine]]
        for machine in comparison.index
    }

    comparison['revised_sequence'] = pd.Series(revised_filtered)
    comparison['changed'] = comparison['original_sequence'] != comparison['revised_sequence']

    return comparison



df_compare = compare_machine_sequences_after_T1(df_plan_init, df_plan_01_bic, 1440)
df_compare
```

```python
def kendall_tau_from_sequences(original_sequence: list, revised_sequence: list) -> tuple[float, float]:
    """
    Berechnet Kendall's Tau zwischen zwei Job-Sequenzen (gleiche Elemente, evtl. andere Reihenfolge).

    Gibt zurück:
    - tau      : Kendall’s Tau-Korrelationskoeffizient
    - p_value  : p-Wert zur statistischen Signifikanz
    """
    # Rangpositionen im Original
    rank_original = {job: i for i, job in enumerate(original_sequence)}

    # Reihenfolge der Originalränge nach revised_sequence
    rank_revised = [rank_original[job] for job in revised_sequence]
    
    # Vergleichsbaseline: 0, 1, 2, ...
    rank_baseline = list(range(len(rank_revised)))
    
    # Tau und p-Wert
    tau, p_value = kendalltau(rank_baseline, rank_revised)
    return tau, p_value
```

```python
seq_original = df_compare.loc['M4', 'original_sequence']
seq_revised  = df_compare.loc['M4', 'revised_sequence']

tau, p = kendall_tau_from_sequences(seq_original, seq_revised)
print(f"Kendall's Tau für M0: {tau:.3f} (p-Wert: {p:.4f})")
```

```python
def find_order_violations(original_sequence: list, revised_sequence: list) -> list:
    """
    Gibt alle Jobpaare (a, b) zurück, die im Original a < b hatten, aber in der Revision b < a.

    Rückgabe:
    - offenders : Liste von Tupeln (a, b), deren Reihenfolge verletzt wurde
    """
    rank_revised = {job: i for i, job in enumerate(revised_sequence)}
    offenders = []

    for i in range(len(original_sequence)):
        for j in range(i + 1, len(original_sequence)):
            a = original_sequence[i]
            b = original_sequence[j]
            if rank_revised[a] > rank_revised[b]:
                offenders.append((a, b))

    return offenders

offenders = find_order_violations(seq_original, seq_revised)
print(f"Verletzte Paare: {len(offenders)}")
for a, b in offenders:
    print(f"  {a} ↔ {b}")
```

```python
def spearman_footrule(seq1: list, seq2: list) -> int:
    """
    Berechnet Spearman's Fußregel-Distanz zwischen zwei Job-Sequenzen.
    Beide Sequenzen müssen die gleichen Elemente enthalten.
    """
    # Mapping von Job → Rang in beiden Sequenzen
    rank1 = {job: i for i, job in enumerate(seq1)}
    rank2 = {job: i for i, job in enumerate(seq2)}

    # Summe der absoluten Differenzen der Ränge
    return sum(abs(rank1[job] - rank2[job]) for job in seq1)
```

```python
footrule = spearman_footrule(seq_original, seq_revised)
print(f"Spearman’s Fußregel für M0: {footrule}")
```

```python
n = len(seq_original)
max_footrule = (n**2) // 2 if n % 2 == 0 else (n**2 - 1) // 2
normalized = footrule / max_footrule
print(f"Normierter Fußregelwert: {normalized:.3f}")
```
